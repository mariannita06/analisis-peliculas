```python
import pandas as pd

df = pd.read_csv("peliculas.csv", sep=';')

# Ver primeras filas
df.head()
```

```python
print(df.shape)
```

```python
# Eliminar filas sin título o rating
df = df.dropna(subset=["title", "rating"])

# Convertir rating a número
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

# Eliminar nulos otra vez
df = df.dropna(subset=["rating"])

df.head()
```

```python
promedio_genero = df.groupby("genre")["rating"].mean().reset_index()
promedio_genero = promedio_genero.sort_values(by="rating", ascending=False)

print(promedio_genero)
```

```python
top10 = df[["title", "rating"]].sort_values(by="rating", ascending=False).head(10)

print(top10)
```
